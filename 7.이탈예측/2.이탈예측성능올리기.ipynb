```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import random
from sklearn.model_selection import train_test_split
```

```python
# 데이터 살펴보기
# 유저 ID - 이탈여부
df = pd.read_csv('train_v2_sample.csv')
df.sample(10)
```

```python
# 유저 ID - 유저 정보
df_1 = pd.read_csv('members_v3_sample.csv', dtype={'gender':'category', 'registered_via':'category','city':'category'})
df_1.head()
```

```python
# 유저 ID - 거래내역
df_2 = pd.read_csv('transactions_v2.csv', dtype={'payment_method_id' : 'category'})
df_2.head()
```

```python
# df 와 df_1을 합친다
total_df = pd.merge(left=df,right=df_1,how='left',on='msno')
total_df.head()
```

```python
# df_2에서 데이터에서 대한 정보를 가져온다
all_df = pd.merge(left=total_df,right=df_2,how='left',on='msno')
all_df.head()
```

```python
len(all_df)
```

```python
# 유저 별 거래 통계량 구하기
stats = all_df.groupby('msno').agg(
        {'msno' : 'count', 'plan_list_price': 'sum', 'actual_amount_paid': 'mean'}
    ).rename(
        columns={'msno': 'total_order', 'plan_list_price': 'plan_net_worth', 'actual_amount_paid': 'mean_payment_each_transaction'}
        ).reset_index()

stats.head()
```

```python
# 유저 정보에 거래 이력 통계 정보를 합치기
fin_df = pd.merge(left=total_df, right=stats, how='left', on='msno')
fin_df.head(10)
```

```python
# 이상치 확인
fin_df['bd'].min(), fin_df['bd'].max()
```

```python
# 아웃라이어 제거
fin_df['bd'].clip(0,100, inplace=True) 
fin_df['bd'].fillna(0,inplace=True)

# 제거 확인
fin_df['bd'].min(), fin_df['bd'].max()
```

```python
# 필요없는 컬럼, 결측값 제거
fin_df.drop(columns='msno', inplace=True)
fin_df.dropna(inplace=True)
```

```python
# 더미화
fin_df_dummies = pd.get_dummies(fin_df)
fin_df_dummies.head()
```

```python
X = fin_df_dummies.drop(columns='is_churn')
y = fin_df_dummies['is_churn'].values
```

```python
# 학습 / 테스트 데이터
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=0)
```

## 트리 앙상블 모델로 이탈 예측

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# 모델 초기화 및 학습하기
tree_model = DecisionTreeClassifier(random_state=0, max_depth=2)
tree_model.fit(train_X, train_y)
```

```python
# 모델 평가하기 (정확도)
from sklearn import metrics
prediction_test = tree_model.predict(test_X)
print(metrics.f1_score(test_y, prediction_test))
```

#### 랜덤포레스트 모델로 예측

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# 모델 초기화 및 학습하기
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(train_X, train_y)
```

```python
# 모델 평가하기 (정확도)
from sklearn import metrics
prediction_test = rf_model.predict(test_X)
print(metrics.f1_score(test_y, prediction_test))
```
